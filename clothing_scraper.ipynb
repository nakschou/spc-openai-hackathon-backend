````python
#create a function that parsed a given html component and returns a 2d array, where there is a link for the item in the first entry, and a link for the image in the second entry
def get_desc(img_url, clothing_item):
    response = client.chat.completions.create(
    model="gpt-4-vision-preview",
    messages=[
        {
        "role": "user",
        "content": [
            {"type": "text", "text": "Look for the following clothing item: " + clothing_item + "and give a two sentence description of the item. Make \
                sure to include information about the color, material, and any other important features."},
            {
            "type": "image_url",
            "image_url": {
                "url": img_url,
            },
            },
        ],
        }
    ],
    max_tokens=300,
    )

    return response.choices[0].message.content

def get_embedding(text, model="text-embedding-3-small"):
    text = text.replace("\n", " ")
    return client.embeddings.create(input = [text], model=model).data[0].embedding


def get_links(soups, clothing_item):
    # Find the image tag and extract the src attribute
# Find all article tags within the div
    articles = soups.find('div').find_all('article')

    # Initialize the master list to store extracted links
    master_list = []

    for article in articles:
        img_tag = soups.find('img', {'name': 'product-module-image'})
        image_src = img_tag['src'] if img_tag else None

        # Find the anchor tag and extract the href attribute
        anchor_tag = soups.find('article').find('a', href=True)
        item_href = anchor_tag['href'] if anchor_tag else None

        desc = get_desc(image_src, clothing_item)
        embedding = get_embedding(desc)


        # Append the extracted links to the master list
        master_list.append([desc, image_src, 'https://www.nordstromrack.com' + item_href, embedding])

    return master_list
````
